### 8. <u>Treating Missing Data<u>

##### 8.1 Replacing NaN 'Gender' with "Unknown"

```python
df.loc[df['Gender'].isna(),'Gender']='Unknown'
df['Gender']
```

##### 8.2 Dropping 'Dependents' having NaN values

```python
df = df.dropna(subset=['Dependents'])
```

##### 8.3 Replacing NaN 'Self_Employed' with "Unknown"

```python
df.loc[df['Self_Employed'].isna(),'Self_Employed']='Unknown'
df['Self_Employed']
```

##### 8.4 Replacing NaN 'LoanAmount' with "mean"

```python
df['LoanAmount'] = np.where(df['LoanAmount'].isna(),df['LoanAmount'].mean(), df['LoanAmount'])
df['LoanAmount']
```

##### 8.5 Replacing NaN 'Loan_Amount_Term' with "median"

```python
df['Loan_Amount_Term'] = np.where(df['Loan_Amount_Term'].isna(),df['Loan_Amount_Term'].median(), df['Loan_Amount_Term'])
df['Loan_Amount_Term']
```

##### 8.6 Replacing NaN 'Credit_History' with "median"

```python
df['Credit_History'] = np.where(df['Credit_History'].isna(),df['Credit_History'].median(), df['Credit_History'])
df['Credit_History']
```

### 9. <u>Treating Outliers<u>

```python
def replace_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace values below lower bound with lower bound and values above upper bound with upper bound
    data[column] = data[column].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    
    return data

data = replace_outliers(df, 'ApplicantIncome')
```

```python
data = replace_outliers(df, 'CoapplicantIncome')
```

```python
data = replace_outliers(df, 'LoanAmount')
```
